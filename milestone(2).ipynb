i must now create a function to make all meanings of never buying from the store as a one unique value

```python
def never_value (enjoy_method):
  if enjoy_method in ['Never', 'Never buy', 'never']:
    return 0
  elif enjoy_method == 'I dont like coffee':
    return 1
  elif enjoy_method == 'Drive-thru' :
    return 2
  elif enjoy_method == 'Dine in' :
    return 3
  elif enjoy_method == 'Take away' :
    return 4

df['enjoy_method'] =df.apply(lambda x :never_value(x.enjoy_method), axis=1 )

```

```python
def loyality(visit_frequency,enjoy_method,visit_duration,membership_card):
  if visit_frequency ==0 and enjoy_method ==0 :
    return 0 # non_customer
  elif visit_frequency == 4 :
    return 4 #high loyal customer
  elif visit_frequency ==3  and visit_duration ==1 :
    return 3 # loyal customer
  elif visit_frequency ==1 and membership_card==0 :
    return 1 #not loyal customer
  else :
    return 2 #normal custmer
df["loyal"] = df.apply(lambda x : loyality(x.visit_frequency,x.enjoy_method,x.visit_duration,x.membership_card),axis=1)
df
```

```python
def revenue_potential(row):
    #  weights for importnat thigs
    visit_weight = 1.5
    duration_weight = 1
    spend_weight = 2
    quality_weight = 2
    continue_buying_weight = 2

    # clc potential score
    potential_score = (
        row['visit_frequency'] * visit_weight +
        row['visit_duration'] * duration_weight +
        row['spend_per_visit'] * spend_weight +
        row['quality_rating'] * quality_weight +
        row['continue_buying'] * continue_buying_weight
    )


    if potential_score >= 15:
        return 3  # High Potential
    elif 10 <= potential_score < 15:
        return 2  # Moderate Potential
    else:
        return 1  # Low Potential


df['revenue_potential'] = df.apply(revenue_potential, axis=1)
```
