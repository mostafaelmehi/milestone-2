```python
def loyality(visit_frequency,enjoy_method,visit_duration,membership_card):
  if visit_frequency ==0 or enjoy_method ==0 :
    return 0 # non_customer
  elif visit_frequency == 4 :
    return 4 #high loyal customer
  elif visit_frequency ==3  and visit_duration ==1 :
    return 3 # loyal customer
  elif visit_frequency ==1 and membership_card==0 :
    return 1 #not loyal customer
  else :
    return 2 #normal custmer
df["loyal"] = df.apply(lambda x : loyality(x.visit_frequency,x.enjoy_method,x.visit_duration,x.membership_card),axis=1)
df
```

we are trying to create a wight to ranl our customers more specifc

```python
def revenue_potential(row):
    # Updated weights based on relative importance
    visit_weight = 1.0  # Reduced to reflect lesser importance compared to spend
    duration_weight = 1.0  # Kept the same as duration is moderately important
    spend_weight = 3.0  # Increased to emphasize the critical role of spending
    quality_weight = 1.5  # Included as a moderate factor
    continue_buying_weight = 2.0  # Kept to reflect long-term retention value
    loyal = 2.0  # Maintained as loyalty remains crucial

    # Calculate the potential score
    potential_score = (
        row['visit_frequency'] * visit_weight +
        row['visit_duration'] * duration_weight +
        row['spend_per_visit'] * spend_weight +
        row['continue_buying'] * continue_buying_weight +
        row['loyal'] * loyal
    )

    # Adjusted thresholds based on potential score distribution
    if potential_score >= 16:
        return 2  # High Potential
    elif 8 <= potential_score < 16:
        return 1  # Moderate Potential
    else:
        return 0  # Low Potential

# Apply the updated function to classify customers
df['SpendingPotential'] = df.apply(revenue_potential, axis=1)

```
